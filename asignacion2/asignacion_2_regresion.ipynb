- Los datos están organizados para que puedas ver la evolución por trimestre y por año, lo que permite análisis de tendencias.
- Se pueden usar variables financieras como predictors (por ejemplo: activos, pasivos, ingresos) y variables objetivo como beneficio, rentabilidad, crecimiento, etc.

## Documentación de las columnas del dataset

**Balance General Anual de Empresas Públicas**

### Columnas de identificación:
- **stock**: Símbolo bursátil de la empresa (string). Ejemplo: "AAPL", "MSFT". Identifica únicamente cada empresa.
- **endDate**: Fecha de cierre del periodo fiscal (YYYY-MM-DD). Indica el último día del año fiscal (anual).

### Activos:
- **cash**: Efectivo y equivalentes de efectivo (float, USD). Representa el dinero disponible inmediatamente.
- **shortTermInvestments**: Inversiones a corto plazo (float, USD). Inversiones líquidas de menos de 1 año.
- **netReceivables**: Cuentas por cobrar netas (float, USD). Dinero que deben los clientes menos cuentas incobrables.
- **inventory**: Inventario (float, USD). Valor de mercancías y materias primas disponibles para venta.
- **otherCurrentAssets**: Otros activos corrientes (float, USD). Activos líquidos menores no categorizados.
- **totalCurrentAssets**: Activos corrientes totales (float, USD). Suma de todos los activos convertibles a efectivo en < 1 año.
- **longTermInvestments**: Inversiones a largo plazo (float, USD). Inversiones con horizonte > 1 año.
- **propertyPlantEquipment**: Propiedad, planta y equipo (float, USD). Activos físicos fijos (PPE).
- **goodWill**: Fondo de comercio (float, USD). Valor intangible de adquisiciones sobre valor contable.
- **intangibleAssets**: Activos intangibles (float, USD). Patentes, marcas, licencias, etc.
- **otherAssets**: Otros activos (float, USD). Activos no corrientes no clasificados.
- **deferredLongTermAssetCharges**: Cargos diferidos de activos a largo plazo (float, USD). Gastos capitalizados amortizables.
- **totalAssets**: Activos totales (float, USD). Suma de todos los activos de la empresa.
- **netTangibleAssets**: Activos tangibles netos (float, USD). Activos totales - intangibles - pasivos.

### Pasivos:
- **accountsPayable**: Cuentas por pagar (float, USD). Dinero que la empresa debe a proveedores.
- **otherCurrentLiab**: Otros pasivos corrientes (float, USD). Obligaciones a corto plazo no especificadas.
- **totalCurrentLiabilities**: Pasivos corrientes totales (float, USD). Suma de deudas pagaderas en < 1 año.
- **shortLongTermDebt**: Deuda a corto/largo plazo (float, USD). Porción de deuda a largo plazo vencida en < 1 año.
- **longTermDebt**: Deuda a largo plazo (float, USD). Obligaciones de deuda con vencimiento > 1 año.
- **deferredLongTermLiab**: Pasivos diferidos a largo plazo (float, USD). Obligaciones pospuestas (impuestos, pensiones).
- **otherLiab**: Otros pasivos (float, USD). Pasivos no corrientes no clasificados.
- **totalLiab**: Pasivos totales (float, USD). Suma de todas las obligaciones de la empresa.

### Capital contable:
- **commonStock**: Acciones comunes (float, USD). Valor nominal de acciones ordinarias emitidas.
- **capitalSurplus**: Superávit de capital (float, USD). Capital pagado en exceso del valor nominal.
- **retainedEarnings**: Utilidades retenidas (float, USD). Ganancias acumuladas no distribuidas como dividendos.
- **treasuryStock**: Acciones de tesorería (float, USD). Acciones recompradas por la empresa (negativo).
- **otherStockholderEquity**: Otro capital accionario (float, USD). Componentes de capital no categorizados.
- **totalStockholderEquity**: Capital contable total (float, USD). Activos totales - pasivos totales.
- **minorityInterest**: Participación minoritaria (float, USD). Capital de accionistas no controladores en subsidiarias.

**Observaciones generales:**
- Periodicidad: Anual
- Unidad: USD (dólares estadounidenses)
- Valores NaN indican que la empresa no reportó ese dato para el periodo
- El dataset contiene 17,511 registros de 4,400+ empresas públicas
- Periodo aproximado: últimos 4 años por empresa

# Paso 5 — División 80/20 (entrenamiento / prueba)

Objetivo: separar los datos en 80% para entrenamiento y 20% para prueba respetando la estructura temporal por empresa (sin hacer shuffle).

Recomendaciones:
- Asegúrate de tener las columnas `stock`, `endDate` y `cash` y que el dataframe esté ordenado por `stock` y `endDate` (ascendente).
- Para series temporales por empresa, usa una división basada en tiempo: los primeros 80% de observaciones de cada `stock` → train; los últimos 20% → test.



````python
# Paso 5: División temporal 80/20 por empresa
import math

# Diccionario para almacenar los conjuntos de entrenamiento y prueba por empresa
train_dict = {}
test_dict = {}

print("=" * 70)
print("DIVISIÓN TEMPORAL 80/20 POR EMPRESA (Sin Shuffle)")
print("=" * 70)

for stock in top_3_stocks:
    # Filtrar datos de la empresa y ordenar por fecha
    df_stock = df_final[df_final['stock'] == stock].sort_values('endDate').reset_index(drop=True)
    
    # Calcular el índice de separación (80% para train)
    n_total = len(df_stock)
    n_train = math.floor(n_total * 0.8)  # Redondear hacia abajo
    n_test = n_total - n_train
    
    # Dividir respetando el orden temporal
    train = df_stock.iloc[:n_train].copy()
    test = df_stock.iloc[n_train:].copy()
    
    # Almacenar en diccionarios
    train_dict[stock] = train
    test_dict[stock] = test
    
    # Mostrar información de la división
    print(f"\n📊 Empresa: {stock}")
    print(f"   Total de observaciones: {n_total}")
    print(f"   • Train (80%): {n_train} observaciones")
    print(f"     └─ Periodo: {train['endDate'].min().date()} a {train['endDate'].max().date()}")
    print(f"   • Test (20%): {n_test} observación(es)")
    print(f"     └─ Periodo: {test['endDate'].min().date()} a {test['endDate'].max().date()}")

# Concatenar todos los conjuntos para análisis global
df_train_all = pd.concat([train_dict[stock] for stock in top_3_stocks], ignore_index=True)
df_test_all = pd.concat([test_dict[stock] for stock in top_3_stocks], ignore_index=True)

print("\n" + "=" * 70)
print("RESUMEN GLOBAL:")
print(f"  • Total Train: {len(df_train_all)} registros")
print(f"  • Total Test: {len(df_test_all)} registros")
print(f"  • Proporción: {len(df_train_all)/len(df_final)*100:.1f}% train / {len(df_test_all)/len(df_final)*100:.1f}% test")
print("=" * 70)

print("\n⚠️ IMPORTANTE: No se aplicó shuffle para preservar el orden temporal")
print("   En series temporales, el modelo debe predecir el futuro basándose en el pasado.")

print("\n✅ Datos listos para entrenamiento del modelo de regresión lineal")

# Mostrar ejemplo de train/test para una empresa
print(f"\n📋 Ejemplo de división para {top_3_stocks[0]}:")
print("\nConjunto de ENTRENAMIENTO:")
display(train_dict[top_3_stocks[0]])
print("\nConjunto de PRUEBA:")
display(test_dict[top_3_stocks[0]])
````

# Paso 6 — crea y entrena el modelo de regresión lineal para predecir `cash` por empresa

Objetivo: entrenar un modelo de regresión lineal por cada `stock` usando la serie temporal (respetando orden temporal) y evaluar en el 20% final.

Pasos recomendados:
- Preprocesamiento
    - Definir variables para el entrenamiento: y = `cash` (variable objetivo). `data` → X (conjunto de características a usar para predecir `cash`, p. ej. rezagos de `cash`, indicadores temporales, variables exógenas). Usar X e y en el entrenamiento: `model.fit(X_train, y_train)`.
    - Asegurar `endDate` como datetime y ordenar por `stock`, `endDate`.  - Asegurar `endDate` como datetime y ordenar por `stock`, `endDate`.



````python
# Paso 6: Crear y entrenar modelo de regresión lineal por empresa
from sklearn.linear_model import LinearRegression

# Diccionario para almacenar los modelos entrenados
models = {}
X_train_dict = {}
y_train_dict = {}
X_test_dict = {}
y_test_dict = {}

print("=" * 70)
print("ENTRENAMIENTO DE MODELOS DE REGRESIÓN LINEAL POR EMPRESA")
print("=" * 70)

for stock in top_3_stocks:
    print(f"\n📈 Entrenando modelo para: {stock}")
    print("-" * 70)
    
    # Obtener conjuntos de train y test
    train = train_dict[stock].copy()
    test = test_dict[stock].copy()
    
    # --- CREAR FEATURES TEMPORALES ---
    # Como solo tenemos la fecha, usaremos días desde la fecha inicial como feature
    # Esto permite que el modelo aprenda la relación temporal
    fecha_inicial = train['endDate'].min()
    
    # Para TRAIN: calcular días desde la fecha inicial
    train['dias_desde_inicio'] = (train['endDate'] - fecha_inicial).dt.days
    
    # Para TEST: calcular días desde la misma fecha inicial
    test['dias_desde_inicio'] = (test['endDate'] - fecha_inicial).dt.days
    
    # Definir X (features) e y (variable objetivo)
    X_train = train[['dias_desde_inicio']].values
    y_train = train['cash'].values
    
    X_test = test[['dias_desde_inicio']].values
    y_test = test['cash'].values
    
    # Almacenar para uso posterior
    X_train_dict[stock] = X_train
    y_train_dict[stock] = y_train
    X_test_dict[stock] = X_test
    y_test_dict[stock] = y_test
    
    # --- ENTRENAR MODELO ---
    model = LinearRegression()
    model.fit(X_train, y_train)
    
    # Almacenar modelo entrenado
    models[stock] = model
    
    # --- INFORMACIÓN DEL MODELO ---
    print(f"✅ Modelo entrenado exitosamente")
    print(f"   • Feature utilizada: 'dias_desde_inicio' (tiempo)")
    print(f"   • Observaciones de entrenamiento: {len(X_train)}")
    print(f"   • Rango temporal train: {train['endDate'].min().date()} a {train['endDate'].max().date()}")
    print(f"\n   📊 Parámetros del modelo:")
    print(f"   • Coeficiente (pendiente): {model.coef_[0]:,.2f}")
    print(f"   • Intercepto: ${model.intercept_:,.2f}")
    print(f"   • Ecuación: cash = {model.intercept_:,.2f} + {model.coef_[0]:,.2f} * dias")
    
    # Interpretar el coeficiente
    if model.coef_[0] > 0:
        print(f"   • Interpretación: El efectivo aumenta ${model.coef_[0]:,.2f} por día")
    else:
        print(f"   • Interpretación: El efectivo disminuye ${abs(model.coef_[0]):,.2f} por día")
    
    # Predicciones en train (para verificar ajuste)
    y_train_pred = model.predict(X_train)
    train_score = model.score(X_train, y_train)
    print(f"   • R² en Train: {train_score:.4f}")

print("\n" + "=" * 70)
print("✅ TODOS LOS MODELOS ENTRENADOS EXITOSAMENTE")
print("=" * 70)
print(f"\nModelos disponibles: {list(models.keys())}")
print(f"Total de modelos: {len(models)}")
print("\n🎯 Los modelos están listos para hacer predicciones en el conjunto de prueba")
````

- **Simular escenario real de predicción:** En aplicaciones prácticas, entrenamos con datos históricos (pasado) y predecimos periodos futuros. La división temporal 80/20 replica este flujo: el modelo aprende de 2016-2018 y debe predecir 2019 sin haberlo visto antes.

- **Validar capacidad de generalización temporal:** Queremos evaluar si el modelo puede extrapolar patrones históricos hacia el futuro, no solo interpolar dentro del rango conocido.

**Consecuencias de barajar:**

- Métricas irrealmente optimistas (el modelo ve patrones futuros durante el entrenamiento)
- Pérdida de autocorrelaciones temporales
- Modelo inútil para predicción real (no sabríamos si generaliza al futuro verdadero)

**Conclusión:** En series temporales, el tiempo define la estructura natural de los datos. Respetar el orden temporal es esencial para entrenar modelos que realmente puedan predecir el futuro con datos del pasado.

### 3. ¿Cuál es la diferencia entre MSE y RMSE y qué nos dice cada métrica sobre la precisión de las predicciones?

**Respuesta:**

**Definiciones matemáticas:**

- **MSE (Mean Squared Error - Error Cuadrático Medio):**
  
  $$\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$

  Promedio de los errores al cuadrado entre valores reales $(y_i)$ y predichos $(\hat{y}_i)$.

- **RMSE (Root Mean Squared Error - Raíz del Error Cuadrático Medio):**
  
  $$\text{RMSE} = \sqrt{\text{MSE}} = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}$$

  Es simplemente la raíz cuadrada del MSE.

**Diferencias clave:**

| Aspecto | MSE | RMSE |
|---------|-----|------|
| **Unidades** | Unidades al cuadrado (USD²) | Mismas unidades que la variable objetivo (USD) |
| **Interpretabilidad** | Menos intuitivo (valores muy grandes) | Más intuitivo (error promedio en USD) |
| **Sensibilidad a outliers** | Altamente sensible (errores grandes se amplifican exponencialmente) | Altamente sensible (heredado del MSE) |
| **Uso en optimización** | Preferido en matemática (función convexa, derivable) | Usado para reporting y comunicación |

**¿Qué nos dice cada métrica?**

**MSE:**
- Penaliza fuertemente errores grandes debido al cuadrado
- En nuestro caso: ZYXI tiene MSE = $1.33 trillones (USD²), lo cual es difícil de interpretar directamente
- Útil para comparaciones relativas entre modelos (menor MSE = mejor)
- Función de pérdida estándar en algoritmos de optimización

**RMSE:**
- **Interpretación directa:** Representa el error promedio de predicción en las unidades originales (USD)
- Para ZYXI: RMSE = $1,154,333.33 significa que, en promedio, el modelo se equivoca por ~$1.15 millones
- Permite contexto relativo: un RMSE de $1.15M sobre un valor real de $14.04M representa un error del 8.22%
- Más adecuado para comunicar resultados a stakeholders no técnicos

**En nuestro análisis:**

- **ZYXI:** RMSE = $1,154,333.33 → Error del 8.22% → EXCELENTE predicción
- **A:** RMSE = $980,666,666.67 → Error del 70.96% → NECESITA MEJORA
- **ZIXI:** RMSE = $16,161,333.33 → Error del 121.07% → NECESITA MEJORA (predicho valor negativo para cash positivo)

**Conclusión:** El RMSE es preferible para interpretar la precisión del modelo porque está en unidades comprensibles (dólares), mientras que el MSE es más útil para optimización matemática durante el entrenamiento.

### 4. ¿Qué interpreta el valor de R² en este problema (predicción de `cash`) y qué limitaciones tiene su interpretación en series temporales?

**Respuesta:**

**Definición de R² (Coeficiente de Determinación):**

$$R^2 = 1 - \frac{\sum_{i=1}^{n}(y_i - \hat{y}_i)^2}{\sum_{i=1}^{n}(y_i - \bar{y})^2} = 1 - \frac{SS_{res}}{SS_{tot}}$$

Donde:
- $SS_{res}$ = Suma de cuadrados residuales (error del modelo)
- $SS_{tot}$ = Suma de cuadrados totales (varianza de los datos)
- $\bar{y}$ = Media de los valores reales

**Interpretación general:**

- **R² cercano a 1:** El modelo explica la mayor parte de la variabilidad de `cash`. Predicciones muy ajustadas a los valores reales.
- **R² cercano a 0:** El modelo no explica la variabilidad. Las predicciones no son mejores que usar la media como predicción.
- **R² negativo:** El modelo es peor que predecir siempre la media (muy mal ajuste).

**En nuestro problema específico:**

Durante el **entrenamiento**, observamos R² muy altos:
- ZYXI: R² = 0.9981 en train (99.81% de la varianza explicada) → Ajuste cuasi-perfecto en datos históricos
- Resultados similares para A y ZIXI en train

Sin embargo, en el **conjunto de prueba**, obtuvimos **R² = N/A (NaN)** para las 3 empresas.

**¿Por qué R² no es aplicable en nuestro test?**

R² requiere calcular la varianza de los datos reales:

$$\text{Varianza} = \frac{1}{n}\sum_{i=1}^{n}(y_i - \bar{y})^2$$

**Con solo 1 observación de test por empresa:**
- No se puede calcular una media significativa
- La varianza es cero (no hay dispersión con un solo dato)
- La fórmula de R² resulta en división por cero → NaN

**Limitaciones de R² en series temporales:**

1. **Sensibilidad a tendencias:** 
   - R² puede ser artificialmente alto en series con tendencias fuertes (ascendente/descendente), incluso si las predicciones individuales son malas.
   - Un modelo naive que simplemente siga la tendencia puede obtener R² alto sin capturar patrones reales.

2. **No detecta autocorrelación:**
   - R² no identifica si el modelo captura correctamente las dependencias temporales (autocorrelación, estacionalidad).
   - Puede ignorar errores sistemáticos en la estructura temporal.

3. **Problemas con extrapolación:**
   - Un R² alto en train no garantiza buena predicción futura (test).
   - En nuestro caso: R² > 0.99 en train, pero errores del 8% a 121% en test.
   - Esto sugiere sobreajuste o incapacidad del modelo lineal simple para capturar patrones complejos.

4. **Inadecuado para muestras pequeñas:**
   - Como vimos, con 1 sola observación de test, R² no es calculable.
   - Incluso con pocas observaciones, R² es inestable y poco confiable.

5. **No considera el orden temporal:**
   - R² trata todas las observaciones como independientes, ignorando la secuencia temporal.
   - No penaliza errores que violan patrones temporales.

**Métricas alternativas recomendadas para series temporales:**

- **MAE/RMSE:** Errores absolutos más interpretables y robustos.
- **MAPE (Mean Absolute Percentage Error):** Error porcentual promedio (ya lo usamos: 8.22%, 70.96%, 121.07%).
- **AIC/BIC:** Para comparar modelos considerando complejidad.
- **Validación walk-forward:** Evaluar en múltiples ventanas temporales progresivas.
- **Residuos autocorrelacionados (ACF/PACF):** Verificar si los errores tienen estructura temporal residual.

**Conclusión:** En nuestro problema, R² no fue útil en test (N/A por muestra única) y en train puede ser engañoso (alto pero no garantiza generalización). Para series temporales con datos limitados, **es preferible usar RMSE, MAE y error porcentual**, que proporcionan información más relevante sobre la capacidad predictiva real del modelo.
